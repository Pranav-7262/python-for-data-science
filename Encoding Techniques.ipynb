# Encoding techniques

### Label Encoding

```python
import pandas as pd

df = pd.DataFrame({

    'Employee_ID':[101,102,103,104,105,106,107,108,109,110],

    'Gender':[
        'Male','Female','Male','Female','Male',
        'Female','Male','Female','Male','Female'
    ],

    'Department':[
        'HR','IT','Finance','IT','Marketing',
        'HR','Finance','Marketing','IT','HR'
    ],

    'Education_Level':[
        'High School',
        'Bachelors',
        'Masters',
        'PhD',
        'Bachelors',
        'Masters',
        'PhD',
        'High School',
        'Bachelors',
        'Masters'
    ],

    'City':[
        'Mumbai',
        'Delhi',
        'Pune',
        'Mumbai',
        'Bangalore',
        'Delhi',
        'Chennai',
        'Pune',
        'Mumbai',
        'Bangalore'
    ],

    'Performance':[
        'Good',
        'Excellent',
        'Average',
        'Excellent',
        'Good',
        'Average',
        'Excellent',
        'Good',
        'Excellent',
        'Average'
    ]
})

df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
df["Gender"].value_counts()
```

```python
df["Gender"] = LE.fit_transform(df["Gender"]) #used when there is a two unique values in a column
```

```python
df["Gender"][:3]
```

```python
df["Gender"].value_counts() #used to count the value
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order = OrdinalEncoder()
```

```python
df["Performance"] = order.fit_transform(df[["Performance"]]) # this gives default values to the column values
```

```python
df["Performance"]
```

```python
# to give specific encoded values to the column
oe = OrdinalEncoder(
    categories = [[        
        'Good',
        'Average',
        'Excellent'
    ]]
)
df["Performance"] = oe.fit_transform(df[["Performance"]])
print(df)
```

```python
df["Education_Level"].value_counts()
```

```python
df["Education_Level"] = order.fit_transform(df[["Education_Level"]])
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(
    categories=[[
        'High School',
        'Bachelors',
        'Masters',
        'PhD'
    ]]
)

df["Education_Level"] = oe.fit_transform(df[["Education_Level"]])

df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(
    categories=[[
        'Bangalore',
        'Chennai',
        'Delhi',
        'Mumbai',
        'Pune'
    ]]
)

df["City"] = oe.fit_transform(df[["City"]])

print(df.head())
```
