# Regular Expression

```python
import re
```

```python
text = "Regex session starts soon and NLP class begins."
```

```python
res = re.findall("[A-Za-z.]{1,10}" , text)
res
```

```python
text = """My Mobile Number is 9967590895"""
res = re.findall("[A-Za-z0-9]{1,10}" , text)
res
```

```python
res = re.findall('[a-zA-Z]{2,10}' , text)
res
```

```python
t = "My PIN is 416001 and I am from Kolhapur!"
res = re.findall('[a-zA-Z0-9!]{1,10}' , t)
res
```

```python
text = """My Mobile Number is 9967590895
        my email id is : chaskarakshay9597@gmail.com
        today's date is : 18/5/2025
        """
no = re.findall('[0-9]{10}',text)
no
```

```python
email = re.findall('[a-z]{1,30}[0-9]{4}[@]{1}[a-z.]{4,10}.',text)
email
```

```python
date = re.findall('[0-9]{2}[/][0-9/]{0,6}',text)
date
```

```python
text = """My Mobile Number is 996  7590  895
        my email id is : chaskarakshay%$#9597@gmail.com
        today's date is : "18-5-2025"
        """
r1 = re.findall('[0-9]{3}[  ]+[0-9]{4}[  ]+[0-9]{3}',text)
r1
```

```python
text = """
        1234 245 0987    
        124 235 098     
        4567 2345 0987123     
        123 234 087    
        """
res = re.findall('\d{1,9}\s\d{1,9}\s\d{1,9}',text)
res
```

```python
 # Find valid ISFC code    (dont use /b)
text = """ Bank IFSC Code is :
            HDFC0009873
            HDFC0034098733456789
            9899887766
            989988445533
            
            ASDFG5678P
            
            BFDFG5566PQWERTYUI
            
            RTYQWERT4567POIUY"""
res = re.findall(r'[A-Z]{5}[0-9]{4}[A-Z]\b',text)
res
```
