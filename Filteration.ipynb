```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='Windows-1252')
```

```python
df
```

# Filter DF based on Condition

```python
df['Company'] == 'Apple'
```

```python
df[df['Company'] == 'Apple']
```

```python
df[df['Company'] == 'Apple'].value_counts('Company')
```

```python
# Find which row not have "HP" in a Company column
df[df['Company'] != 'HP'].value_counts('Company')
```

```python
# Find laptops over 2000 euros
df[df['Price_euros'] > 2000.0]
```

# Create Column based on Condition

```python
import numpy as np
```

```python
tier = np.where(df['Price_euros'] > 2000.0 , "Expensive" , "Cheap")
tier
```

```python
df['Product_tier'] = tier
df.head()
```

```python
df[df['Product_tier'] == "Cheap"].value_counts("Product_tier")
```

# Filter based on multiple condition

```python
# find apple laptops that costs more than 2000
df[(df['Company'] == 'Apple') & (df['Price_euros'] > 2000.0)]
```

```python
# find laptops whic are apple or dells
df[(df['Company'] == 'Apple') | (df['Company'] == 'Dell')]
```

```python
# find laptops of apple or dell that cost over 2000
df[((df['Company'] == 'Apple') | (df['Company'] == 'Dell')) & (df['Price_euros'] > 2000.0)]
```
