# Organize all data in dictionary

```python
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {} 
for league in dict_countries:
    list=[]
    for season in range(15, 21):
        dataframe = pd.read_csv(root + str(season)+ str(season+1) + '/' + dict_countries[league] + '.csv')
        dataframe.insert(1 , 'season' , season)
        list.append(dataframe)
    df_frames = pd.concat(list) # storing all seasons of a league in a single dataframe
    dict_historical_data[league] = df_frames # assigning dataframe to a new dictionary
```
