# With Arrays 1.Numpy


```python
import pandas as pd
import numpy as np
data = np.array([ # create an array
    [1,4],
    [2,5],
    [3,6]
])
```

```python
# creating dataframe
df = pd.DataFrame(data , index = ['r1','r2','r3'] , columns = ['c1','c2'])
```

```python
df
```

# Using Dictionary

```python
countries = ["india","USA","China","Brazil","England"]
pops = [1500000000,347275807,1700000000,110000000,10000000]
```

```python
dict = {'Countries' : countries , 'Population' : pops}
dict
```

```python
df = pd.DataFrame(dict)
df
```
