```python
remove_cols = [
    "temp",
    "humidity",
    "msl_pressure",
    "incidence_angle",
    "zenith",
    "azimuth"
]

df_clean = df.copy()

for col in remove_cols:

    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)

    IQR = Q3 - Q1

    lowerBound = Q1 - (1.5 * IQR)
    upperBound = Q3 + (1.5 * IQR)

    df_clean = df_clean[
        (df_clean[col] >= lowerBound) &
        (df_clean[col] <= upperBound)
    ]

print("Original Shape :", df.shape)
print("Clean Shape :", df_clean.shape)
```

```python
x = df_new.drop("power_kw", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.65)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.intercept_
```

```python
coef_df = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': LR.coef_
})

coef_df = coef_df.sort_values(
    by='Coefficient',
    ascending=False
)

coef_df
```

```python
prediction = LR.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
```

```python
result1 = mean_absolute_error(y_test,prediction)
result1
```

```python
result2 = mean_squared_error(y_test,prediction)
result2
```
