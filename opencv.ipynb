```python

# define a video capture object
vid = cv2.VideoCapture(0)

while(True):

    # Capture the video frame
    # by frame
    ret, frame = vid.read()

    # Display the resulting frame
    cv2.imshow('frame', frame)

    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```

# Image filter

```python
img = cv2.imread(r"C:\Users\dell\Downloads\dog.jpg")
```

```python
if img is None:
    print("Could not read img")
kernel1 = np.array([[2,0,-1],
                   [3,0,-1],
                   [-2,0,-2]])
identity = cv2.filter2D(src=img , ddepth=-1,kernel= kernel1)

cv2.imshow("original",img)
cv2.imshow("Identity",identity)

cv2.waitKey(1000)
cv2.imwrite("identity.jpg" , identity)
cv2.destroyAllWindows()
```

```python
kernel2 = np.ones((5,5) , np.float32)/45
img2 = cv2.filter2D(src=img , ddepth=-1,kernel= kernel2)
cv2.imshow("original",img)
cv2.imshow("Identity",img2)

cv2.waitKey()
cv2.imwrite("blue_kernel.jpg" , img2)
cv2.destroyAllWindows()
```
