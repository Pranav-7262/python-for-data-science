```python
import cv2
```

```python
img = cv2.imread(r"C:\Users\dell\Downloads\mahirat.jpg")
```

```python
t_lower = 50
t_upper = 200
```

```python
edge = cv2.Canny(img , t_lower , t_upper) # Canny age filter
edge
```

# Show original and edges image

```python
cv2.imshow("original", img)
cv2.imshow("edge" , edge)
```
