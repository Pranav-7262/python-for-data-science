# Functions

```python
def sum(a, b):
    return a + b
```

```python
sum(3,4)
```

```python
def fact(n):
    facto = 1
    for i in range(1 ,n+1):
       facto = facto * i
     
    return facto
```

```python
print(fact(5))
```

```python
def factRec(n):
    if n < 2:
        return 1
    return n * factRec(n-1)
```

```python
print(factRec(5))
```
