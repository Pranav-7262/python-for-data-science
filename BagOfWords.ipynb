```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
corpus1 = ["This is a sample sentence" , "This is a another another is example"]
```

```python
v1 = CountVectorizer() ## convert text into bow representation
x1 = v1.fit_transform(corpus1)
x1
```

```python
print("Vocabilary: ", v1.get_feature_names_out()) # unique words in corpus
```

```python
print("BOW representation: \n",x1.toarray()) # BOW matrix
```

```python
v2 = CountVectorizer(stop_words='english')
x2 = v2.fit_transform(corpus1)
```

```python
print("Vocabilary: ", v2.get_feature_names_out()) # unique words in corpus
```

```python
print("BOW representation: \n",x2.toarray()) # BOW matrix
```

```python
v3 = CountVectorizer(ngram_range=(2,2)) # capture word pairs
x3 = v3.fit_transform(corpus1)
print("Vocabilary: ", v3.get_feature_names_out()) # unique words in corpus
print("BOW representation: \n",x3.toarray()) # BOW matrix
```

```python
texts = ["this product is good", "this product is bad"]

vec = CountVectorizer(max_df=2)  # max_df removes terms that appear in too many documents
X = vec.fit_transform(texts)

print(pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out()))
```

```python
texts = ["this product is good", "this product is bad"]

vec = CountVectorizer(min_df=2)  # max_df removes rare values
X = vec.fit_transform(texts)

print(pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out()))
```

```python
texts = ["this product is good", "this product is bad"]

vec = CountVectorizer(max_features=6)  # limits vocabulary size to the top N terms.
X = vec.fit_transform(texts)

print(pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out()))
```
