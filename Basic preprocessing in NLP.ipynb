### 5.1 Stemming

```python
from nltk.stem import LancasterStemmer
import re

lancaster = LancasterStemmer()

text = """NLTK does not provide Marathi stopwords by default. We typically create a custom Marathi stopword list or use publicly available Marathi stopword datasets. After tokenization, we filter tokens that appear in the stopword list before applying stemming, lemmatization, or feature extraction techniques such as TF-IDF."""

words = re.findall(r'\b\w+\b',text)

stemmed_words = [lancaster.stem(i) for i in words]

print("Original words: ",words)
print("Stemmed words: ",stemmed_words)
```

### 5.2 Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

#nltk.download('wordnet')
#nltk.download('punkt')

lemmatizer = WordNetLemmatizer()

sentence = "NLTK does not provide Marathi stopwords by default. We typically create a custom Marathi stopword list or use publicly available Marathi stopword datasets. After tokenization, we filter tokens that appear in the stopword list before applying stemming, lemmatization, or feature extraction techniques such as TF-IDF."

words = word_tokenize(sentence.lower())

lemmatized_words = [lemmatizer.lemmatize(word,pos="v") for word in words]

print("\n Original Sentence: \n\n",sentence)
print("\n Stemmed Sentence: \n\n", " ".join(stemmed_words))
print("\n Lemmatized Sentence: \n\n", " ".join(lemmatized_words))

```
